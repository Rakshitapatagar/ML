```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv('/content/drive/MyDrive/linear_regression.csv')
print(df)
a=df['x'].to_numpy().reshape(-1,1)
b=df['y'].to_numpy()
print(a)
print(b)
plt.scatter(a,b)
plt.xlabel("IV")
plt.ylabel("DV")
plt.show()
from sklearn.linear_model import LinearRegression
SLR=LinearRegression()
SLR.fit(a,b)
pred=SLR.predict(a)
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(b,pred)
plt.plot(a,pred,color='Red',marker='*')
print("MSE:",mse)
print("intercept:",SLR.intercept_)
print("slope:",SLR.coef_)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Read the CSV file into a DataFrame
data = pd.read_csv('/content/drive/MyDrive/multivariate.csv')

# Extract the independent variables (features) and the dependent variable (target)
x = data[['x1', 'x2', 'x3']].to_numpy()  # Independent variables
y = data['y'].to_numpy()                  # Dependent variable

# Create a Linear Regression model
lr = LinearRegression()

# Fit the model to the data
lr.fit(x, y)

# Predict the target variable using the model
y_pred = lr.predict(x)

# Calculate the mean squared error
mse = mean_squared_error(y, y_pred)

# Print the results
print("Mean Squared Error:", mse)
print("Intercept:", lr.intercept_)
print("Coefficients:", lr.coef_)

# Scatter plot of x1 vs. y
plt.scatter(data['x1'], y)
plt.xlabel('x1')
plt.ylabel('y')
plt.show()

# Calculate and print the correlation matrix between x1, x2, x3, and y
correlation_matrix = data[['x1', 'x2', 'x3', 'y']].corr()
print("Correlation Matrix:")
print(correlation_matrix)
```

```python
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error

# Generate random data
random.seed(42)
tp = [random.random() for _ in range(10)]
k = [random.random() for _ in range(10)]

x = np.array(tp).reshape((-1, 1))
y = np.array(k)
# Create polynomial features
degree = 2 # You can change this to the degree you want
poly = PolynomialFeatures(degree)
x_poly = poly.fit_transform(x)

# Fit the polynomial regression model
poly_lr = LinearRegression()
poly_lr.fit(x_poly, y)
pre_poly = poly_lr.predict(x_poly)

# Calculate mean squared error
mse_poly = mean_squared_error(y, pre_poly)

# Plot the results
plt.scatter(x, y, label="Data")
plt.plot(x, pre_poly, color='red', marker='*', label='Polynomial Prediction')
plt.xlabel("IV")
plt.ylabel("DV")
plt.legend()
plt.show()

print("Polynomial Regression:")
print("MSE:", mse_poly)
print("Intercept:", poly_lr.intercept_)
print("Coefficients:", poly_lr.coef_)
```
